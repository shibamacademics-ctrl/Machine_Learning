```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Insurance.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Premium'],marker = '+',color = 'red')
plt.xlabel("Age")
plt.ylabel("Premium")
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['Age']],df['Premium'],test_size = 0.2,random_state = 20)
model1 = LinearRegression()
```

```python
Linear = model1.fit(x_train,y_train)
Linear
```

```python
Linear.predict(x_test)
```

```python
Linear.score(x_test,y_test)
```

```python
model2 = LogisticRegression()
Logistic = model2.fit(x_train,y_train)
Logistic
```

```python
Logistic.predict(x_test)
```

```python
Logistic.score(x_test,y_test)
```
