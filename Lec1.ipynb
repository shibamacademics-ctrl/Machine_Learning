```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("ML1.csv")
df
```

```python
#Scatter plot
%matplotlib inline
plt.xlabel("area(sq ft)")
plt.ylabel("US $")
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,df.price,color = "blue")
```

```python
#Single linear regression y=ax + b, a=slope/coefficient b=intercept
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#Similarly
reg.predict([[5000]])
```
