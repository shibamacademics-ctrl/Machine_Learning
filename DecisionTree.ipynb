```python
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import pandas as pd
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop("salary_more_then_100k", axis = "columns" )
target = df["salary_more_then_100k"]
inputs.head()
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()

inputs["company_n"] = le_company.fit_transform(inputs['company'])
inputs["Job_n"] = le_job.fit_transform(inputs['job'])
inputs["degree_n"] = le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
final_df = pd.concat([inputs_n,target],axis = 1)
final_df.head()
```

```python
model = tree.DecisionTreeClassifier()
y_pred = model.fit(inputs_n,target)
y_pred
```

```python
y_pred.predict([[2,2,1]])
```

```python
model.score(inputs_n,target)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
inputs = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis = 'columns')
target = df['Survived']
inputs.head()
```

```python
target.head()
```

```python
inputs['Fare'] = inputs['Fare'].fillna(0).astype(int)
inputs['Age'] = inputs['Age'].fillna(0).astype(int)
le_Sex = LabelEncoder()
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex']) 
inputs_n = inputs.drop('Sex',axis = 'columns')
inputs_n.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs_n,target,test_size = 0.2,random_state = 42)
model1 = LinearRegression()
model2 = LogisticRegression()
model1.fit(x_train,y_train)
```

```python
model2.fit(x_train,y_train)
```

```python
model1.predict(x_test)
```

```python
model2.predict(x_test)
```

```python
model1.score(x_test,y_test)
```

```python
model2.score(x_test,y_test)
```

```python
model3 = tree.DecisionTreeClassifier()
model3.fit(x_train,y_train)
model3
```

```python
model3.predict(x_test)
```

```python
model3.score(x_test,y_test)
```

```python
final_df = pd.concat([inputs_n,target],axis = 1)
final_df.head()
```
