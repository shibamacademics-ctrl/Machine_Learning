```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
df = pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df = df[cols_to_use]
df.head()
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)

# other continuous features can be imputed with mean for faster results since our focus is on Reducing overfitting
# using Lasso and Ridge Regression
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
df.head()
```

```python
df.Method.unique()
```

```python
df = df.drop('Suburb',axis = 'columns')
le = LabelEncoder()
df['Type'] = le.fit_transform(df.Type)
df['Method'] = le.fit_transform(df.Method)
df.head()
```

```python
df['Price'] = df['Price'].fillna(df.Price.mean())
df.head()
```

```python
print(df.dtypes)
```

```python
df = pd.get_dummies(df,drop_first = False,dtype = int)
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
df.head()
```

```python
input = df.drop('Price',axis='columns')
target = df.Price
x_train,x_test,y_train,y_test = train_test_split(input,target,test_size = 0.2)
model = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model1=linear_model.Ridge(alpha=50, max_iter=100, tol=0.1)
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model1.score(x_train,y_train)
```
